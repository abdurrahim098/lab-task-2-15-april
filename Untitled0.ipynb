PART 1

```python

class Trapezoid:
    def __init__(self, base1, base2, height):
        self.base1 = base1
        self.base2 = base2
        self.height = height

    def area(self):
        return 0.5 * (self.base1 + self.base2) * self.height

class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height
```

PART 2


```python
def compare_areas(shape1, shape2):
    if shape1.area() > shape2.area():
        return "First shape has larger area"
    elif shape1.area() < shape2.area():
        return "Second shape has larger area"
    else:
        return "Both shapes have equal area"

```

Let's demonstrate the `compare_areas` function with some examples.

```python
trapezoid1 = Trapezoid(base1=5, base2=7, height=4)
parallelogram1 = Parallelogram(base=6, height=4)

print(f"Trapezoid Area: {trapezoid1.area()}")
print(f"Parallelogram Area: {parallelogram1.area()}")

result1 = compare_areas(trapezoid1, parallelogram1)
print(f"Comparison Result 1: {result1}")

trapezoid2 = Trapezoid(base1=4, base2=6, height=5)
parallelogram2 = Parallelogram(base=5, height=5)

print(f"\nTrapezoid Area: {trapezoid2.area()}")
print(f"Parallelogram Area: {parallelogram2.area()}")

result2 = compare_areas(trapezoid2, parallelogram2)
print(f"Comparison Result 2: {result2}")
```

part 3


```python
areas = {}
t = Trapezoid(10, 6, 5)
p = Parallelogram(8, 5)

areas["Trapezoid"] = t.area()
areas["Parallelogram"] = p.area()

print("Areas dictionary:", areas)

```

part 4

```python
dimensions = [
    (Trapezoid(8, 6, 4), Parallelogram(7, 5)),
    (Trapezoid(12, 10, 6), Parallelogram(9, 8)),
    (Trapezoid(5, 5, 3), Parallelogram(4, 4))
]

for t, p in dimensions:
    print("Trapezoid area:", t.area(), "Parallelogram area:", p.area())


```

```python
max_shape = max(areas, key=areas.get)
print("Shape with maximum area:", max_shape, "=", areas[max_shape])
```
